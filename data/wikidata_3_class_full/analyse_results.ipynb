````python
prop = lambda x: [i for i in x.split(' ') if 'prop' in i or 'wdt' in i][0] if len([i for i in x.split(' ') if 'prop' in i or 'wdt' in i]) > 0 else None
ent = lambda x: [i for i in x.split(' ') if 'entity' in i][0] if len([i for i in x.split(' ') if 'entity' in i]) > 0 else None
````

````python
sample_df_tp_e = sample_df_tp
sample_df_tp_e['p'] = sample_df_tp['triple_pattern'].apply(prop)
sample_df_tp_e['e'] = sample_df_tp['triple_pattern'].apply(ent)
sample_df_tp_e
````

````python
ents = list(set(sample_df_tp_e['e']))
preds = list(set(sample_df_tp_e['p']))
````

````python
def df_filter_ents_pred(ents, preds, df):
    es = None
    for e in ents:
        if e is not None:
            if es is None:
                es = df['queryString'].str.contains(e)
            else:
                es = es | df['queryString'].str.contains(e)
    ps = None
    for p in preds:
        if p is not None:
            if ps is None:
                ps = df['queryString'].str.contains(p)
            else:
                ps = ps | df['queryString'].str.contains(p)
    return es, ps, es|ps
es, ps, comb = df_filter_ents_pred(ents, preds, test_sampled)
````

````python
train_sampled = pd.read_csv("/data/wikidata_3_class_full/train_sampled.tsv", sep='\t')
val_sampled = pd.read_csv("/data/wikidata_3_class_full/val_sampled.tsv", sep='\t')
````

````python
for i in [train_sampled, val_sampled]:
    es, ps, comb = df_filter_ents_pred(ents, preds, i)
    print(len(i), len(i[es]), len(i[ps]), len(i[ps|es]))
````

````python
def affected(affected, total):
    return (affected/total)*100
````

````python
for i in [train_sampled, val_sampled, test_sampled]:
    es, ps, comb = df_filter_ents_pred(ents, preds, i)
    print('affected in % ', affected(len(i[ps|es]),len(i)))
````

````python
def df_filter_tp(sample_df_tp_e, df):
    filt = pd.Series([False for x in range(len(df))])
    for idx, row in sample_df_tp_e.iterrows():
        if row['e'] == None:
            filt = filt | df['queryString'].str.contains(row['p'])
        else:
            filt = filt | (df['queryString'].str.contains(row['p']) & df['queryString'].str.contains(row['e']))
    return filt

for i in [train_sampled, val_sampled, test_sampled]:
    filt = df_filter_tp(sample_df_tp_e, i)
    print('affected in % ', affected(len(i[filt]),len(i)), "    ", len(i[filt]))
````

````python

def df_filter_tp_gen(sample_df_tp_e, df):
    for idx, row in sample_df_tp_e.iterrows():
        if row['e'] == None:
            yield df['queryString'].str.contains(row['p']), row['p']
        else:
            yield (df['queryString'].str.contains(row['p']) & df['queryString'].str.contains(row['e'])), row['p']+"   "+row['e']

specific_tp_test_log = {'Triple Pattern':[], 'Affected of QL (%)': []}
for i in [test_sampled]:
    gen = df_filter_tp_gen(sample_df_tp_e, i)
    for filt, trpl in gen:
        specific_tp_test_log['Triple Pattern'].append(trpl)
        specific_tp_test_log['Affected of QL (%)'].append(affected(len(i[filt]),len(i)))
        print(trpl,'     affected in % ', affected(len(i[filt]),len(i)), "    ", len(i[filt]))
````
