```python
from graph_construction.query_graph import snap_lat2onehotv2
import numpy as np
```

```python
int3_ids = int_3_all['id']
list(int3_ids)
```

```python
int_6_all['time_cls'] = int_6_all['mean_latency'].apply(lambda x: np.argmax(snap_lat2onehotv2(x)))
```

```python
#int_6_all_filt = 
int_6_all_filt= int_6_all[~int_6_all['id'].isin(int3_ids)]
int_6_all_filt
```

```python
slow_unseen_qs = int_6_all_filt[int_6_all_filt['time_cls']==2]
med_unseen_qs = int_6_all_filt[int_6_all_filt['time_cls']==1]
len(med_unseen_qs),len(slow_unseen_qs)
```

```python
slow_unseen_qs_wo_distinct= slow_unseen_qs[~slow_unseen_qs['query_string_0'].str.contains('DISTINCT')][['id','query_string_0', 'mean_latency']]
```

```python

for lat, idx, qstr in zip(list(slow_unseen_qs_wo_distinct['mean_latency']),list(slow_unseen_qs_wo_distinct['id']),list(slow_unseen_qs_wo_distinct['query_string_0'])):
    print(lat,'  ',idx, '     ',qstr,'\n')
```
